```python
#Import 
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
import pickle
```

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
import datetime
```

```python
with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)

with open('one_hot_encoder.pkl','rb') as file:
    one_hot_encoder  = pickle.load(file)

with open('scalar.pkl','rb') as file:
    scaler = pickle.load(file)
model = load_model('model.h5')
```

```python
one_hot_encoder.get_feature_names_out()
```

```python
input_data = {
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000

}
```

```python
ont_hot_geo=one_hot_encoder.transform([[input_data['Geography']]]).toarray()
```

```python
one_hot_encoder.get_feature_names_out(['Geography'])
```

```python
label_gender=label_encoder_gender.transform([[input_data['Gender']]])
```

```python
df_geo = pd.DataFrame(ont_hot_geo,columns=one_hot_encoder.get_feature_names_out(['Geography']),index=None)
df_geo
```

```python
input_df = pd.DataFrame([input_data])
```

```python
input_df
```

```python
input_df['Gender']=label_gender
```

```python
input_df
```

```python
input_df = pd.concat([input_df.drop('Geography',axis=1),df_geo],axis=1)
```

```python
input_df
```

```python
## scaling the data
input_scale = scaler.transform(input_df)
input_scale
```

```python
## Prediction
prediction = model.predict(input_scale)
prediction
```

```python
prediction_proba = prediction[0][0]
```

```python
if(prediction_proba>0.5):
    print("exited")
else:
    print('nOT Excited')
```
